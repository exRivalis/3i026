## Rappels sur les TPs précédent

```python
class LabeledSet:  
    
    def __init__(self, input_dimension):
        self.input_dimension = input_dimension
        self.nb_examples = 0
    
    def addExample(self, vector, label):
        if (self.nb_examples == 0):
            self.x = np.array([vector])
            self.y = np.array([label])
        else:
            self.x = np.vstack((self.x,vector))
            self.y = np.vstack((self.y,label))
        
        self.nb_examples = self.nb_examples + 1
    
    #Renvoie la dimension de l'espace d'entrée
    def getInputDimension(self):
        return self.input_dimension
        
    #Renvoie le nombre d'exemples dans le set
    def size(self):
        return self.nb_examples
    
    #Renvoie la valeur de x_i
    def getX(self, i):
        return self.x[i]
            
    #Renvoie la valeur de y_i
    def getY(self, i):
        return self.y[i]
```

```python
#dessin ligne de separation
def plot_frontiere(set, classifier, step=20):
    mmax = set.x.max(0)
    mmin = set.x.min(0)
    x1grid, x2grid = np.meshgrid(np.linspace(mmin[0], mmax[0], step), np.linspace(mmin[1], mmax[1], step))
    grid = np.hstack((x1grid.reshape(x1grid.size, 1), x2grid.reshape(x2grid.size, 1)))
    
    # calcul de la prediction pour chaque point de la grille
    res = np.array([classifier.predict(grid[i,:]) for i in range(len(grid)) ])
    res = res.reshape(x1grid.shape)
    # tracé des frontieres
    plt.contourf(x1grid, x2grid, res, colors=["red", "cyan"], levels=[-1000,0,1000], linewidth=2)

#dessin des points
def plot2DSet(dataset):
    plt.scatter(dataset.x[0:int(dataset.size()/2),0],dataset.x[0:int(dataset.size()/2),1],marker='o')
    plt.scatter(dataset.x[int(dataset.size()/2):dataset.size(),0],dataset.x[int(dataset.size()/2):dataset.size(),1],marker='x')
    
#separe un dataset en deux datasets
# p valeur entre 0 et 100
def split(the_set, p):
    #creation d'une liste en partant du dataset
    list = [(the_set.getX(i), the_set.getY(i)[0]) for i in range(the_set.size())]
    lim = int(len(list)*1.*p/100)
    #print ('limit', lim)
    list1 = list[0:lim]
    list2 = list[lim:]
    
    #les deux datasets de sortie
    set1 = LabeledSet(the_set.getInputDimension())
    set2 = LabeledSet(the_set.getInputDimension())
    
    for l in list1:
        set1.addExample(l[0], l[1])
    for l in list2:
        set2.addExample(l[0], l[1])
        
    return set1, set2
```

## Étape 1 : Perceptron de Rosenblatt


En héritant de la classe `Classifier`, implanter le perceptron tel que décrit en cours. La fonction `train` fera une itération sur l'ensemble des données de l'apprentissage. Les points d'apprentissage seront tirés dans un ordre aléatoire.

```python
class Perceptron(Classifier):
    
    def __init__(self, input_dimension, learning_rate, nombre_iterations):
        self.input_dimension = input_dimension
        self.learning_rate = learning_rate
        self.nombre_iterations = nombre_iterations
        self.w = np.random.randn(self.input_dimension)*self.input_dimension
        
    def train(self, dataset):
        #ici on optimise la valeur de w en fonction des donnees d'entrainnement
        #cette fonction renvoie la liste de levolution de la precision
        precision = []
        #pour chaque (input, output) on modifie legerement w 
        #split 80-20 -> train sur 80% et valider sur 20%
        ds1, ds2 = split(dataset, 80)
        #appretissage sur 80% (ds1)
        for j in range(self.nombre_iterations):
            for i in range(ds1.size()):
                x = dataset.getX(i)
                y = dataset.getY(i)[0]
                if y*self.predict(x) <= 0:
                    for k in range(self.input_dimension):
                        self.w[k] += y*x[k]*self.learning_rate
            #toutes les 10 iterations on evalue nos modifications
           
            if j%10 == 0:
                precision.append((j, self.accuracy(ds2)))
                acc = self.accuracy(ds2)
                #print ("accuracy ", acc)
            
        #validation sur 20% (ds2)
        #acc = self.accuracy(ds2)
        
        return precision 
        
    def predict(self, x):
        # si <x|w> >= 0 "+1", sinon "-1"
        #p = x[0]*self.w[0] + x[1]*self.w[1]
        return np.sign(np.dot(x, self.w))
        
    
    def accuracy(self, dataset):
        #pour chaque x, y dans dataset comparer predict(x) et y
        #predict depend du modèle 
        cpt = 0;
        for i in range(dataset.size()):
            if self.predict(dataset.getX(i)) == dataset.getY(i):
                cpt+=1
        return float(cpt) / dataset.size()
```

La méthode `train` devra faire un nombre d'itérations fixées à la création de l'objet.

Pour pouvoir évaluer la qualité de l'apprentissage au cours des itérations, on évaluera régulièrement (par exemple toutes les 10 itérations) l'accuracy du modèle, sur un ensemble de validation (un sous-ensemble de l'ensemble d'apprentissage). 

Après chaque apprentissage, on veillera à tracer la courbe d'apprentissage, c'est à dire le score obtenu au cours des itérations, et à tracer la frontière de décision.
