## La moyenne

### Exercice: 
Ecrire la fonction moyenne(valeurs) qui calcule la moyenne des éléments d'un tableau (en utilisant la boucle ''for'', sans utiliser les fonctions déjà implémentées dans numpy)

```python
def moyenne(tab):
    s = 0;
    for v in tab:
        s += v
    return s/len(tab)

moyenne(prices_pd["MedQ"].values)
```

```python
## Densité cumulée

### Exercice: 
Calculez la densité cumulée à partir de prix_moyens, avec un intervalle de discrétisation donné et dessinez. La méthode doit retourner deux tableau: l'abscisse X (les prix moyens possibles entre prix min et prix max), et la densité associée
```

```python
def densite(tab, size):
    interval = pd.cut(tab, size)
    effectif = pd.value_counts(interval)
    dnsty = [float(e)/size for e in effectif]
    d_c = dnsty
    for i in range(1, len(d_c)):
        d_c[i] += d_c[i-1]
     
    
    #l = [min(tab) + i*20 for i in range(int((max(tab) - min(tab)))/size)]
    #print l
    return np.unique(interval), d_c

l, d = densite(prix_moyens, 20)

p = pd.DataFrame(d)
p.plot(kind='kde')#plt.hist(d, bins=20)
```

### Exercice : 
Ecrire la fonction Q(alpha,x,y) qui permet de trouver le quartile(alpha,x,y) - quartile(0.5,x,densite) correspondra à la médiane. Calculez et dessinez (sur une feuille) le boxplot

```python
def Q(alpha, x, y):
    """
        alpha = 0.5 pour mediane
        alpha = [.25, .50 , .75, 1.] pour quartile
    """
    q = pd.qcut(y,  4)
    
    #dessiner
    a = pd.DataFrame(y)
    a.boxplot()
    #print np.unique(q)
    return (np.unique(q))[int(alpha*4)-1]

x, y = densite(prix_moyens, 20)

q = Q(0.5, x, y)
print q
```

## Variance

### Exercice : 
On souhaite maintenant rajouter une colonne ''HighQ_var'' aux données originales contenant la variance des prix par état. ATTENTION, cela suppose de traiter les états (State) les uns après les autres

```python
def variance(l):
    #la somme des pow((x - moy(X)), 2) / n
    x_mean = moyenne(l)
    s =0
    for x in l:
        s += pow(x - x_mean, 2)
    return s / len(l)
        
    """states_var = df.groupby("State")["HighQ"].var()
    var = []
    for s in df.State:
        #var.append()
        var.append(states_var[s])
    """
    
#df['HighQ_var'] = var

#variance(prices_pd)
#print prices_pd
v = []
grouped = prices_pd.groupby('State')
for s in prices_pd.State:
    #variance pour chaque etat
    vs = variance(grouped.get_group(s)['HighQ'].values)
    v.append(vs)
    
prices_pd['HighQ_var'] = v

print prices_pd
```

### Exercice

A l'aide de la fonction moyenne écrite précédemment, écrire une fonction correlation(x,y) qui calcule la correlation lineaire entre deux tableaux numpy

```python
"""
    coef de correlation= cov(x, y)/(ecart_type(x) * ecart_type(y))
    ecart_type = sqrt(variance)
"""
def correlation(x, y):    
    c = np.cov([x, y], bias=True)[0][1]#covariance de x et y
    cr = c / (np.sqrt(variance(x)) * np.sqrt(variance(y)))
    
    return cr

correlation(x, y)
```

```python
print("La correlation est :%f"%correlation(prix_ca_ny["HighQ_x"].values,prix_ca_ny["HighQ_y"].values))
#correlation(prix_ca_ny["HighQ_x"].values,prix_ca_ny["HighQ_y"].values)
```
